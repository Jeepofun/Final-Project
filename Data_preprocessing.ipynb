```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/Jeepofun/Final-Project/main/Augmented%20Data.csv')
#df = df.drop(columns=['Unnamed: 0'])
df = df.replace(r'^\s*$', np.nan, regex=True)
df = df.rename(columns={'Base_ Engine': 'Base_Engine'})
df.head(162)
```

```python
df.info()
```

```python
#Impute missings
missing = df.isnull().sum().sort_values(ascending=False)
missing.head()
```

```python
df['Range']=df['Range'].astype('float64')
df['Range'] = df['Range'].fillna(df['Range'].mean())
df['Batt']=df['Batt'].astype('float64')
df['Batt'] = df['Batt'].fillna(df['Batt'].mean())
```

```python
df.isnull().sum()
```

```python
df['Mileage']=df['Mileage'].fillna(df['Mileage'].mean())
```

```python
df['Min_cargo_L']=df['Min_cargo_L'].fillna(df['Min_cargo_L'].min())
```

```python
df.Front_headroom.info()
```

```python
df['Front_headroom'] = pd.to_numeric(df['Front_headroom'])

df['Front_headroom']=df['Front_headroom'].fillna(df['Front_headroom'].mean())
df['Front_legroom']=df['Front_legroom'].astype('float64')
df['Front_legroom']=df['Front_legroom'].fillna(df['Front_legroom'].mean())
df['Rear_legroom']=df['Rear_legroom'].astype('float64')
df['Rear_legroom']=df['Rear_legroom'].fillna(df['Rear_legroom'].mean())
df['Rear_shoulder']=df['Rear_shoulder'].astype('float64')
df['Rear_shoulder']=df['Rear_shoulder'].fillna(df['Rear_shoulder'].mean())
```

```python
df['Base_Power']=df['Base_Power'].fillna(df['Base_Power'].mean())
df['Fuel_capacity']=df['Fuel_capacity'].fillna(df['Fuel_capacity'].mean())
```

```python
df['Towing'] = df['Towing'].fillna(7500.0)
```

```python
df.isnull().sum()
```

```python
#One hot encode categoricals
df = pd.get_dummies(df, columns=['Body_style', 'Base_Engine', 'Drive', 'Seats'])
```
